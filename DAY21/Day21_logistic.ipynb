```python
df2=pd.read_csv('general_data.csv')
```

```python
df2.isnull().sum()
```

```python
df2['NumCompaniesWorked'].fillna(df2['NumCompaniesWorked'].median(),inplace=True)
df2['TotalWorkingYears'].fillna(df2['TotalWorkingYears'].median(),inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
lc=LabelEncoder()
```

```python
df2['Attrition']=lc.fit_transform(df2['Attrition'])
df2['BusinessTravel']=lc.fit_transform(df2['BusinessTravel'])
df2['Department']=lc.fit_transform(df2['Department'])
df2['EducationField']=lc.fit_transform(df2['EducationField'])
df2['Gender']=lc.fit_transform(df2['Gender'])
df2['JobRole']=lc.fit_transform(df2['JobRole'])
df2['MaritalStatus']=lc.fit_transform(df2['MaritalStatus'])
df2['Over18']=lc.fit_transform(df2['Over18'])
```

```python
df2.head()
```

```python
x=df2[['Age',  'BusinessTravel', 'Department', 'DistanceFromHome',
       'Education', 'EducationField', 'EmployeeCount', 'EmployeeID', 'Gender',
       'JobLevel', 'JobRole', 'MaritalStatus', 'MonthlyIncome',
       'NumCompaniesWorked', 'Over18', 'PercentSalaryHike', 'StandardHours',
       'StockOptionLevel', 'TotalWorkingYears', 'TrainingTimesLastYear',
       'YearsAtCompany', 'YearsSinceLastPromotion', 'YearsWithCurrManager']]
y=df2['Attrition']
```

```python
x2=sm.add_constant(x)
```

```python
logistic_attrition=sm.Logit(y,x2)
```

```python
result2=logistic_attrition.fit(method='newton')
```

```python
result2.summary()
```
