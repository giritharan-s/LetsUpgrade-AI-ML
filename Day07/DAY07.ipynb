```python
from sklearn import preprocessing
label_encoder=preprocessing.LabelBinarizer()
df2=df1
df2['Attrition']=label_encoder.fit_transform(df2['Attrition'])
```

```python
df2.head()
```

```python
from scipy.stats import pearsonr
stats,p=pearsonr(df2.Attrition,df2.DistanceFromHome)
print(stats,p)
```

```python
stats,p=pearsonr(df2.Attrition,df2.MonthlyIncome)
print(stats,p)
```

```python
stats,p=pearsonr(df2.Attrition,df2.YearsAtCompany)
print(stats,p)
```

```python
stats,p=pearsonr(df2.Attrition,df2.YearsSinceLastPromotion)
print(stats,p)
```

r = -0.033, there’s low negative correlation between Attrition and yearssincelastpromotion
As the P value of 0.518 is > 0.05, we are accepting H0 and hence there’s no significant correlation
between Attrition & yearssincelastpromotion

```python
from scipy.stats import mannwhitneyu
a1=df1[df1['Attrition']==0]['DistanceFromHome']
a2=df1[df1['Attrition']==1]['DistanceFromHome']
stats,p=mannwhitneyu(a1,a2)
print(stats,p)
```

As the P value of 0.4 is >0.05, the H0 is accepted.
H0: There is no significant differences in the distance from home attrition (Y) and attirition
(N)
Ha: There is significant differences in the distance from home (Y) and attirition (N)

```python
a1=df1[df1['Attrition']==0]['MonthlyIncome']
a2=df1[df1['Attrition']==1]['MonthlyIncome']
stats,p=mannwhitneyu(a1,a2)
print(stats,p)
```

As the P value of 0.053 is >0.05, the H0 is accepted.
H0: There is no significant differences in the MonthlyIncome between attrition (Y) and attirition
(N)
Ha: There is significant differences in the MonthlyIncome between attrition (Y) and attirition (N)

```python
a1=df1[df1['Attrition']==0]['Age']
a2=df1[df1['Attrition']==1]['Age']
stats,p=mannwhitneyu(a1,a2)
print(stats,p)
```

```python
As the P value of 2.9951588479067175e-30 is >0.05, the H0 is rejected and Ha is accepted.
H0: There is no significant differences in the age between attrition (Y) and attirition
(N)
Ha: There is significant differences in the age between attrition (Y) and attirition (N)
```
