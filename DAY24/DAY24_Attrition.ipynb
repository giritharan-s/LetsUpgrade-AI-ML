```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('general_data.csv')
```

```python
df['NumCompaniesWorked'].fillna(df['NumCompaniesWorked'].median(),inplace=True)
df['TotalWorkingYears'].fillna(df['TotalWorkingYears'].median(),inplace=True)
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
lc=LabelEncoder()
```

```python
df['Attrition']=lc.fit_transform(df['Attrition'])
```

```python
df['BusinessTravel']=lc.fit_transform(df['BusinessTravel'])
```

```python
df['Department']=lc.fit_transform(df['Department'])
```

```python
df['EducationField']=lc.fit_transform(df['EducationField'])
```

```python
df['EducationField']=lc.fit_transform(df['EducationField'])
df['Gender']=lc.fit_transform(df['Gender'])
df['JobRole']=lc.fit_transform(df['JobRole'])
df['MaritalStatus']=lc.fit_transform(df['MaritalStatus'])
df['Over18']=lc.fit_transform(df['Over18'])
```

```python
df.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
rc=RandomForestClassifier(oob_score=True)
x=df[['Age',  'BusinessTravel', 'Department', 'DistanceFromHome',
       'Education', 'EducationField', 'EmployeeCount', 'EmployeeID', 'Gender',
       'JobLevel', 'JobRole', 'MaritalStatus', 'MonthlyIncome',
       'NumCompaniesWorked', 'Over18', 'PercentSalaryHike', 'StandardHours',
       'StockOptionLevel', 'TotalWorkingYears', 'TrainingTimesLastYear',
       'YearsAtCompany', 'YearsSinceLastPromotion', 'YearsWithCurrManager']]
y=df['Attrition']
```

```python
rc.fit(x,y)
```

```python
print('oob score;',rc.oob_score_)
```

```python
f=['Age',  'BusinessTravel', 'Department', 'DistanceFromHome',
       'Education', 'EducationField', 'EmployeeCount', 'EmployeeID', 'Gender',
       'JobLevel', 'JobRole', 'MaritalStatus', 'MonthlyIncome',
       'NumCompaniesWorked', 'Over18', 'PercentSalaryHike', 'StandardHours',
       'StockOptionLevel', 'TotalWorkingYears', 'TrainingTimesLastYear',
       'YearsAtCompany', 'YearsSinceLastPromotion', 'YearsWithCurrManager']
for f,imp in zip(f,rc.feature_importances_):
    print(f,imp)
    
```

```python

x=df[['Age','DistanceFromHome','MonthlyIncome','PercentSalaryHike','TotalWorkingYears']]
y=df['Attrition']
```

```python
from sklearn.tree import DecisionTreeClassifier
dc=DecisionTreeClassifier(max_depth=10)
dc.fit(x,y)
```

```python
from IPython.display import Image  
from sklearn.externals.six import StringIO  
from sklearn.tree import export_graphviz
import pydot 

features = list(['Age','DistanceFromHome','MonthlyIncome','PercentSalaryHike','TotalWorkingYears'])
features
```
